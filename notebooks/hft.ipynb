```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the dataset
file_path = "data/data.csv" 
 
df = pd.read_csv(file_path)

# 2. Extract columns cleanly using their EXACT text names
# Let's find the matching Bid name (it should be PRICE_BID_0 based on the FI-2010 standard)
ask_col = 'PRICE_ASK_0'
bid_col = 'PRICE_BID_0' if 'PRICE_BID_0' in df.columns else df.columns[23] # fallback just in case

print(f"Using columns: Ask='{ask_col}', Bid='{bid_col}'")

ask_prices = pd.to_numeric(df[ask_col], errors='coerce')
bid_prices = pd.to_numeric(df[bid_col], errors='coerce')

# 3. Calculate Mid-Price
mid_prices = (ask_prices + bid_prices) / 2

# 4. Take a slice of the first 200 ticks so we see a clean, high-res window
start_tick = 0
end_tick = 200 

time_horizon = np.arange(start_tick, end_tick)
asks_slice = ask_prices.iloc[start_tick:end_tick]
bids_slice = bid_prices.iloc[start_tick:end_tick]
mid_slice = mid_prices.iloc[start_tick:end_tick]

# 5. Plotting
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

ax.plot(time_horizon, asks_slice, color='red', linestyle='--', alpha=0.6, label='Best Ask (Sellers)')
ax.plot(time_horizon, bids_slice, color='green', linestyle='--', alpha=0.6, label='Best Bid (Buyers)')
ax.plot(time_horizon, mid_slice, color='blue', linewidth=2, label='Mid-Price')

ax.set_title("Actual FI-2010 Order Book Price Movements", fontsize=14, fontweight='bold')
ax.set_xlabel("Time (Ticks)", fontsize=12)
ax.set_ylabel("Price Value (Integer Units)", fontsize=12)

# Prevent matplotlib from forcing scientific notation on the axis values
ax.get_yaxis().get_major_formatter().set_useOffset(False)

ax.grid(True, linestyle=':', alpha=0.6)
ax.legend(loc='upper right')

plt.tight_layout()
```
